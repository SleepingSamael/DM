```python
#Python libraries
#Classic,data manipulation
import numpy as np
import pandas as pd
# Plots
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df= pd.read_csv('ODI/ODI-2022-cleaned.csv')
fetures = ['Programme', 'ML', 'IR', 'Statistics', 'Databases', 'Gender', 'Stress-level']
df=df[fetures]
for index, row in df.iterrows():
    if row['Stress-level'] == 'unknown':
        df.drop(index=index)
df.describe()
```

```python
corr = df.corr(method='pearson')
fig, ax = plt.subplots(figsize=(25,15))
sns.heatmap(corr, vmax = 0.9, annot= True,cmap = "Greens",square = True)
plt.title("Correlation matrix of features")
plt.show()
```
